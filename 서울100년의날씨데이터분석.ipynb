```python
#df['일시']컬럼의 타입을 'object'에서 'datetime'으로 변경하기
df['일시'] = pd.to_datetime(df['일시'])
df.info()
```

```python
# df['일시']에서 연도만 분리해서 새로운 컬럼명 '연도'생성하기
df['연도'] = df['일시'].dt.year
df.head()
```

```python
df['월'] = df['일시'].dt.month
df.head()
```

```python
# 결측치가 있는 행 번호 리스트로 확인
nullList = df[df.isnull().any(axis=1)].index.tolist()
```

```python
#15676~16431, 21260, 39758, 41519
print(df.iloc[15676]['일시'])
print(df.iloc[16431]['일시'])
print(df.iloc[21260]['일시'])
print(df.iloc[39758]['일시'])
print(df.iloc[41519]['일시'])
```

```python
# df의 15676~16431 index 모두 삭제
df = df.drop(df.index[15676:16432])
```

```python
df.loc[21260]
```

```python
def fillNull(num):
    for colNum in range(3,6):
        if pd.isna(df.loc[num,df.columns[colNum]]):
            df.loc[num,df.columns[colNum]] = (df.loc[num-1,df.columns[colNum]]+df.loc[num+1,df.columns[colNum]])/2

fillNull(21260)
fillNull(39758)
fillNull(41519)
```

```python
df.info()
```

```python
df.reset_index(drop=True, inplace=True)
df.info()
```

```python
start_year = 1907
end_year = 2025
monthAvg =[]
index = []

for year in range(start_year,end_year):
    avg = []
    if year%4 == 0:
        oneYear = 366
    else:
        oneYear = 365

    year_df = df[df['연도']==year]
    if year_df.shape[0] < oneYear:
        delList = df[df['연도']==year].index
        df.drop(delList, inplace=True)
    else:
        index.append(year)
        for month in range(1,13):
            avg.append(year_df[year_df['월']==month]['평균기온(℃)'].mean())
        monthAvg.append(avg)
        print("{}:{}".format(year,year_df.shape[0]))
```

```python
# monthAvg의 배열 크기 확인하기
print(len(monthAvg))
print(len(monthAvg[0]))
```

```python
columns=['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']
month_df = pd.DataFrame(monthAvg, index=index, columns=columns)
month_df.head()
```

```python
month_df.info()
```

```python
yearAvg = []
for year in month_df.index:
    #print("{}:{}".format(year,month_df.loc[year].mean()))
    yearAvg.append(month_df.loc[year].mean())
yearAvg
```

```python
#matplotlib으로 그래프 plot으로 그리기
import matplotlib.pyplot as plt
plt.plot(index,yearAvg)
```

```python
yearAvg_df = pd.DataFrame(yearAvg, index=index, columns=['연평균기온'])
yearAvg_df.head()
```

```python
yearAvg_df['5년이동평균기온'] = yearAvg_df['연평균기온'].rolling(5).mean()
plt.plot(index,yearAvg_df['5년이동평균기온'])
```

```python
yearAvg_df['10년이동평균기온'] = yearAvg_df['연평균기온'].rolling(10).mean()
plt.plot(index,yearAvg_df['10년이동평균기온'])
```
