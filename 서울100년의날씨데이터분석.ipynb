```python
import os

# 기본이 되는 경로
dataPath = '/content/drive/MyDrive/새싹강동3기구글AI과정/code/DataScience/4_데이터분석예제/data/'
# 경로 문자열을 더할때
csvPath1 = os.path.join(dataPath, 'seoul.csv')
csvPath2 = os.path.join(dataPath, 'seoul2.csv')
csvPath3 = os.path.join(dataPath, 'seoul3.csv')
```

```python
import numpy as np
import pandas as pd

df1 = pd.read_csv(csvPath1, encoding='cp949')
df2 = pd.read_csv(csvPath2, encoding='cp949')
df3 = pd.read_csv(csvPath3, encoding='cp949')
```

```python
df1.columns = ['일시', '지점', '평균기온(℃)', '최저기온(℃)', '최고기온(℃)']
df1['지점명'] = '서울'
df1.head()
```

```python
df1 = df1[['지점', '지점명','일시','평균기온(℃)', '최저기온(℃)', '최고기온(℃)']]
df1.head()
```

```python
df1.columns = ['지점', '지점명','일시','평균기온(℃)', '최저기온(℃)', '최고기온(℃)']
df2.columns = ['지점', '지점명','일시','평균기온(℃)', '최저기온(℃)', '최고기온(℃)']
df3.columns = ['지점', '지점명','일시','평균기온(℃)', '최저기온(℃)', '최고기온(℃)']
```

```python
# df1과 df2, df3를 concat
df = pd.concat([df1, df2, df3])
df.head()
```

```python
df.tail()
```

```python
df.reset_index(drop=True, inplace=True)
df.tail()
```

```python
#df['일시']컬럼의 타입을 'object'에서 'datetime'으로 변경하기
df['일시'] = pd.to_datetime(df['일시'])
df.info()
```

```python
# df['일시']에서 연도만 분리해서 새로운 컬럼명 '연도'생성하기
df['연도'] = df['일시'].dt.year
df.head()
```

```python
df['월'] = df['일시'].dt.month
df.head()
```
